```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import helpers
```

```python
shp_path = "ti_sirgas.shp"
```

```python
fig, ax = plt.subplots(figsize=(8,8))
helpers.plot_rs(ax, alpha=0.5)
helpers.plot_areas(ax, shp_path=shp_path, color='red', alpha=1)
plt.title("Terras indígenas")
plt.grid(linestyle="--", alpha=0.5)

red_patch = mpatches.Patch(color='red', label='áreas')
plt.legend(handles=[red_patch])

plt.savefig("indigena.png", bbox_inches='tight')
plt.show()
```
