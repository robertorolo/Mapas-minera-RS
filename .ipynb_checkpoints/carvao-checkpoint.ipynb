```python
import pandas as pd
import chardet
import matplotlib.pyplot as plt
import helpers
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
with open('carvao.csv', 'rb') as f:
    result = chardet.detect(f.read())  # or readline if the file is large

carvao = pd.read_csv('carvao.csv', encoding=result['encoding'], sep=';')
```

```python
carvao.head()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
helpers.plot_rs(ax, alpha=0.5)
sc = helpers.plot_bubble(ax=ax, lat=carvao['lat'], longi=carvao['long'], value=carvao['reservas totais'])
plt.title("Principais jazidas de carvão")
plt.grid(linestyle="--", alpha=0.5)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
cbar = plt.colorbar(sc, cax=cax)
cbar.set_alpha(1)
cbar.draw_all()
plt.savefig("carvao.png", bbox_inches='tight')
plt.show()
```
